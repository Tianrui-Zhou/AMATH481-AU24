```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x * np.sin(3*x) - np.exp(x)

# Define the derivative of f(x)
def df(x):
    return np.sin(3*x) + 3*x*np.cos(3*x) - np.exp(x)

x = np.linspace(-1.5, 0.5, 500)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x)=x*sin(3x)-exp(x)')
plt.grid(True)
plt.show()
```

```python
max_iter = 100  # Set max iteration
tol = 10**(-6)  # Set tolerance

def newton_raphson(x1):
    '''x1 is the initial guess'''
    x_values = [x1]
    
    for i in range(max_iter):
        x_now = x_values[-1]
        x_new = x_now - f(x_now)/df(x_now)
        x_values.append(x_new)
        
        # Check convergence with f(xn)
        if abs(f(x_now)) < tol:
            break
    
    num_of_iter_newton = i + 1
            
    return x_values, num_of_iter_newton
```

```python
def bisection(a, b):
    '''
    a: the left interval endpoint
    b: the right interval endpoint
    '''
    if f(a) * f(b) < 0: # Opposite sign
        mid_points = []
        
        for i in range(max_iter):
            x_mid = (a + b) / 2. #Calculate the midpoint
            mid_points.append(x_mid)
            
            if abs(f(x_mid)) < tol:
                break
            
            if f(x_mid) * f(a) > 0:
                a = x_mid
            
            else:
                b = x_mid
        
        num_of_iter_bisec = i + 1
    
    else:
        raise ValueError("The endpoints must let the values of the function in opposite sign!")
        
    return mid_points, num_of_iter_bisec
```

```python
x_vals_newton, n_iter_newton = newton_raphson(-1.6)

mid_vals_bisec, n_iter_bisection = bisection(-0.7, -0.4)

A1 = np.array(x_vals_newton)
print(A1)
print()
A2 = np.array(mid_vals_bisec)
print(A2)
print()
A3 = np.array([n_iter_newton, n_iter_bisection])
print(A3)
```
